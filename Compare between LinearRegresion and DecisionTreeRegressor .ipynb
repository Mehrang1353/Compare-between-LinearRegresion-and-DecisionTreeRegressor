chek null attribut

```python
data.isnull().sum()
```

Remove the attribut ehich have 118 null as 259

```python
data_c2=data.drop('كيفيت',axis=1)
data_c2
```

Replace null with mean fanction


```python
data_c3=data_c2.fillna(data_c2.mean(axis=0))
data_c3.info()
```

change int and float to str 


```python
li=['پوشش','وضعيت سفارش','کد تخليه','جنس ورق']
data_c3[li]=data_c3[li].astype(str)
data_c3.info()
```

split input and output


```python
x=data_c3.drop('مقدار سفارش',axis=1)
y=data_c3['مقدار سفارش']
print(x.shape)
print(y.shape)
```

Validation

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=101)
```

select Model and learn

```python

from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)

```

```python
y_pred=lr.predict(x_test)
y_pred
```

Evaluation 


```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
mse=mean_squared_error(y_pred,y_test)
mae=mean_absolute_error(y_pred,y_test)
print ('MSE of LinearRegression is :',mse )
print ('RMSE of LinearRegression is :',np.sqrt(mse ))    
print('MAE of LinearRegression is :',mae)
```

 Model DecisinTreeRegration

```python
from sklearn.tree import DecisionTreeRegressor
dr=DecisionTreeRegressor(criterion='mse',max_depth=8,random_state=101)
dr.fit(x_train,y_train)
y_pred=dr.predict(x_test)
mse=mean_squared_error(y_pred,y_test)
mae=mean_absolute_error(y_pred,y_test)
print ('MSE of dr is :',mse )   
print('MAE of dr is :',mae)
```
